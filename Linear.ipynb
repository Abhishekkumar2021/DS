### What is Linear Regression?
Linear regression is a statistical method for modelling relationship between a dependent variable with a given set of independent variables. It is a parametric technique that assumes a linear relationship between the dependent variable (y) and one or more independent variables (x). The case of one independent variable is called simple linear regression. For more than one independent variable, the process is called multiple linear regression.

### Simple Linear Regression
Simple linear regression is an approach for predicting a quantitative response using a single feature (or “predictor” or “input variable”). For example, predicting a person’s weight as a function of their height. The goal of linear regression is to find the line that best fits the data. The line is called the regression line. The equation of the regression line is:

$
y = \beta_0 + \beta_1 x
$

where $\beta_0$ is the intercept and $\beta_1$ is the slope of the line. The $\beta_0$ and $\beta_1$ values are estimated using the training data. The $\beta_0$ and $\beta_1$ values are called the model coefficients. The $\beta_0$ and $\beta_1$ values are estimated using the training data. The $\beta_0$ and $\beta_1$ values are called the model coefficients.

### How to find the best fit line?
The best fit line is the line that minimizes the sum of the squared residuals. The sum of the squared residuals is the vertical distance between the data points and the regression line. 

### $ \sum_{i=1}^n (y_i - \hat{y}_i)^2 $

where $y_i$ is the observed value and $\hat{y}_i$ is the predicted value. The $\hat{y}_i$ is the value of $y$ that corresponds to the $x_i$ value. The $\hat{y}_i$ is calculated using the regression equation. The $\hat{y}_i$ is calculated using the regression equation.

### How to get $ \beta_0 $ and $ \beta_1? $
The $\beta_0$ and $\beta_1$ values can be calculated using the following equations:
$$
\beta_1 = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}
$$
$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

where $\bar{x}$ is the mean of the $x$ values and $\bar{y}$ is the mean of the $y$ values.








```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
from sklearn.linear_model import LinearRegression
sns.set()
```

```python
dataset = datasets.load_diabetes()
X = dataset.data
y = dataset.target

# print(X[:5, :])
print(X.shape)

plt.plot(X[:, 9], y, '*')
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
print(model.score(X, y))

# Draw the regression line
plt.plot(X[:, 9], y, '*')
plt.plot(X[:, 9], model.predict(X), 'r')
plt.show()
```
