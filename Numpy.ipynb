```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
pie = [0.2, 0.3, 0.1, 0.05, 0.35]
plt.pie(pie, labels=['A', 'B', 'C', 'D', 'E'], autopct='%1.1f%%', shadow=True, startangle=90, explode=[0, 0, 0, 0, 0.1], textprops={'fontsize': 14}, wedgeprops={'linewidth': 2, 'edgecolor': '#00000060'}, center=(0, 0), radius=2)
plt.show()
```

```python
df = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [6, 7, 8, 9, 10], 'C': [11, 12, 13, 14, 15], 'D': [16, 17, 18, 19, 20], 'E': [21, 22, 23, 24, 25]})
df.plot(kind='bar', stacked=True, figsize=(10, 5), colormap='tab20', title='Stacked Bar Chart', xlabel='X', ylabel='Y', fontsize=14, rot=0, grid=True, legend=True, edgecolor='#00000060', linewidth=2)
```
