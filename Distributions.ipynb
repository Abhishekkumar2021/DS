## 1. Discrete
### 1.1 Bernoulli
A Bernoulli distribution is a discrete probability distribution that describes the outcome of a single binary experiment. It has two possible outcomes, "success" and "failure", and the probability of success is the same for each trial. The probability of failure is also the same for each trial, and is equal to 1 minus the probability of success. The Bernoulli distribution is a special case of the binomial distribution, where the number of trials is 1.
Example - A coin is tossed. The probability of getting a head is 0.5 and the probability of getting a tail is 0.5.

#### 1.1.1 Probability Mass Function
$$
f(x) = \begin{cases}
p & \text{if } x = 1 \\
1-p & \text{if } x = 0
\end{cases}
$$

#### 1.1.2 Cumulative Distribution Function
$$
F(x) = \begin{cases}
1 & \text{if } x \geq 1 \\
1-p & \text{if } x < 0
\end{cases}
$$

#### 1.1.3 Mean
$$
% Derivation
\begin{align}
\mu &= E[X] \\
&= 1p + 0(1-p) \\
&= p
\end{align}

$$


#### 1.1.4 Variance
$$

% Derivation
\begin{align}
\sigma^2 &= E[(X-\mu)^2] \\
&= E[X^2 - 2X\mu + \mu^2] \\
&= E[X^2] - 2\mu E[X] + \mu^2 \\
&= p(1-p) - 2p(1-p) + p(1-p) \\
&= p(1-p)
\end{align}
$$

#### 1.1.5 Standard Deviation
$$
\sigma = \sqrt{\sigma^2} = \sqrt{p(1-p)}
$$


### 1.2 Binomial
A binomial distribution is a discrete probability distribution that describes the outcome of a series of independent binary experiments. It has two possible outcomes, "success" and "failure", and the probability of success is the same for each trial. The probability of failure is also the same for each trial, and is equal to 1 minus the probability of success. The binomial distribution is a special case of the hypergeometric distribution, where the number of trials is fixed.

#### 1.2.1 Probability Mass Function
$$
f(x) = \begin{cases}
\binom{n}{x}p^x(1-p)^{n-x} & \text{if } x = 0,1,2,...,n \\
0 & \text{otherwise}
\end{cases}
$$

#### 1.2.2 Cumulative Distribution Function
$$
F(x) = \begin{cases}
1 & \text{if } x \geq n \\
\sum_{i=0}^{x}\binom{n}{i}p^i(1-p)^{n-i} & \text{if } x < n
\end{cases}
$$

#### 1.2.3 Mean
$$
% Derivation
\begin{align}
\mu &= E[X] \\
&= np
\end{align}
$$

#### 1.2.4 Variance
$$
% Derivation
\begin{align}
\sigma^2 &= E[(X-\mu)^2] \\
&= E[X^2 - 2X\mu + \mu^2] \\
&= E[X^2] - 2\mu E[X] + \mu^2 \\
&= np(1-p) - 2np(1-p) + np(1-p) \\
&= np(1-p)
\end{align}
$$

#### 1.2.5 Standard Deviation
$$
\sigma = \sqrt{\sigma^2} = \sqrt{np(1-p)}
$$


```python
# binomial distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Random numbers
numbers = binom.rvs(n=5, p=0.3, size=10)
print("Numbers generated: ", numbers)

# plot 
x = np.arange(0, 5)
y = binom.pmf(x, n=5, p=0.3)
plt.plot(x, y, 'o-')
plt.title('Binomial: n=%i, p=%.2f' % (5, 3))
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes')
plt.show()
```

## 1.3 Poisson
A Poisson distribution is a discrete probability distribution that describes the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant rate and independently of the time since the last event. The Poisson distribution can also be used for the number of events in other specified intervals such as distance, area or volume.

#### 1.3.1 Probability Mass Function
$$
f(x) = \begin{cases}
\frac{\lambda^xe^{-\lambda}}{x!} & \text{if } x = 0,1,2,... \\
0 & \text{otherwise}
\end{cases}
$$

#### 1.3.2 Cumulative Distribution Function
$$
F(x) = \begin{cases}
1 & \text{if } x \geq \infty \\
\sum_{i=0}^{x}\frac{\lambda^xe^{-\lambda}}{x!} & \text{if } x < \infty
\end{cases}
$$

#### 1.3.3 Mean
$$
% Derivation
\begin{align}
\mu &= E[X] \\
&= \lambda
\end{align}
$$

#### 1.3.4 Variance
$$
% Derivation
\begin{align}
\sigma^2 &= E[(X-\mu)^2] \\
&= E[X^2 - 2X\mu + \mu^2] \\
&= E[X^2] - 2\mu E[X] + \mu^2 \\
&= \lambda - 2\lambda\lambda + \lambda^2 \\
&= \lambda
\end{align}
$$

#### 1.3.5 Standard Deviation
$$
\sigma = \sqrt{\sigma^2} = \sqrt{\lambda}
$$


```python
# Poisson distribution
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
import seaborn as sns
sns.set(color_codes=True)

# Random numbers
numbers = stats.poisson.rvs(mu=3, size=10)
print("Numbers generated: ", numbers)

# plot
x = np.arange(0, 10)
y = stats.poisson.pmf(x, mu=3)
plt.plot(x, y, 'o-')
plt.title('Poisson: mu=%.2f' % (3))
plt.xlabel('Number of successes')
plt.ylabel('Probability of successes')
plt.show()
```
